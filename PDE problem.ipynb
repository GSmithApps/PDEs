```julia
function A(n , l)
   return 2/l * (-2*l^2/(n * π) * cos(n * π /2) + 4 * l^2 / (n^2 * π ^2) * sin(n * π /2)) 
end
```

```julia
function approx(n,l,x)
    return sum([A(m,l) * sin(m*π*x/l) for m in 1:n])
end
```

```julia
using Plots
```

```julia
l = 1
xs = range(0, l, 20)
ys = [approx(3,l,x) for x in xs]
plot(xs,ys)
```
